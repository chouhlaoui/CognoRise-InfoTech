We need to check the extreme values to see whether it is normal due to some other informations or it might be misleading data.


```python
test = df[df['job_title'] == 'Lead Data Scientist']
import scipy.stats as stats

# Plot histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(test['salary_in_usd'], bins=20, kde=True)

# Overlay a normal distribution curve
mean = test['salary_in_usd'].mean()
std_dev = test['salary_in_usd'].std()
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
plt.plot(x, stats.norm.pdf(x, mean, std_dev), color='red')

plt.title('Salary Distribution with Normal Curve')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
stats.probplot(test['salary_in_usd'], dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

Here we can conclude that the distribution of the Lead Data Scientist salaries is a normal distribution

```python
out = pd.DataFrame(columns=df.columns)
for job in df['job_title'].unique():
  df_job = df[df['job_title'] == job]
  column = 'salary_in_usd'

  Q1 = df_job[column].quantile(0.25)
  Q3 = df_job[column].quantile(0.75)

  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  outliers = df_job[(df_job[column] < lower_bound) | (df_job[column] > upper_bound)]
  if not outliers.empty:
    out = pd.concat([out, outliers])
out
```

We can consider a few theories here:

* **50% Remote Ratio**: It is feasible to maintain a 50% remote work arrangement when residing in one country and working in another within the same continent (e.g., living in France and working in Luxembourg). However, working on different continents (e.g., living in Europe and working in North America) under the same conditions might **raise some questions**, though it **remains** **feasible**.

* **0% Remote Ratio**: Having a 0% remote work arrangement while living and working on different continents might suggest that these combinations could be unusual or **potentially misrecorded**, as they seem **less logical**. This could indicate a need for **closer examination to ensure data accuracy**.

```python
Corr_Matrix = df.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(Corr_Matrix, annot=True, cmap='coolwarm', center=0)
plt.show()
```

```python
df_corr = pd.DataFrame()
df_corr[['var1', 'var2', 'corr']] = Corr_Matrix.unstack().reset_index()
df_corr = df_corr[df_corr['var1'] != df_corr['var2']].sort_values(by='corr', ascending=False)

df_corr.reset_index(drop=True, inplace=True)
indices_to_drop = []

for i in range(len(df_corr) - 1):
    if (df_corr.loc[i, 'var1'] == df_corr.loc[i + 1, 'var2'] and
        df_corr.loc[i, 'var2'] == df_corr.loc[i + 1, 'var1']):
        indices_to_drop.append(i + 1)

df_corr = df_corr.drop(indices_to_drop).reset_index(drop=True)

df_corr.head(10)
```

Based on the correlation matrix and the preivious table, we observe the following key insights:

1.  **Negative Correlations:**

* Company Size and Salary in USD: There is a notable negative correlation, indicating that as company size increases, the salary in USD tends to decrease.
* Job Title and Company Location: A negative correlation is also observed, suggesting that job titles might influence or be influenced by company location.

2. **Positive Correlations:**
* Company Location and Employee Residence: There is a strong positive correlation between these two variables, highlighting that company location is closely related to where employees reside.

* The correlation between experience level and salary in USD is 0.305962, demonstrating a moderate positive relationship. This indicates that higher experience levels are generally associated with higher salaries.
